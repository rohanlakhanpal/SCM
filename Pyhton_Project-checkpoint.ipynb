```python
from tkinter import *
from tkinter import font as tkFont
import tkinter.scrolledtext as st
import random

print("**************")
print("*                                      *")
print("*    Enter 1 to open Guessing Game   *")
print("**   Enter 2 to open Calculator     **")
print("**   Enter 3 to open Password gen   **")
print("*    Enter 0 to exit the program     *")
print("*                                      *")
print("**************")    
while True:
    # Get user input
    choice = input("Enter 1, 2, 3 to open required program, or 0 to exit: ")
    
    # Open the corresponding program
    if choice == "1":
        # Open first program
        # Create the root window
        anmol_root = Tk()
        anmol_root.configure(bg='black')
        anmol_root.resizable(0,0)
        anmol_root.title("Guess the Number")

        # Set the font and text size for the labels and buttons
        font = ('Helvetica', 12, 'bold')

        # Create the labels
        Label(anmol_root, text='Let\'s start the game', bg='black', fg='white', font=font).grid(row=0, column=1)
        Label(anmol_root, text='You will get 10 chances to win the game', bg='black', fg='white', font=font).grid(row=2, column=1)
        Label(anmol_root, text='Enter a number between 0-100', bg='black', fg='white', font=font).grid(row=3, column=1)

        # Create the entry field
        e = Entry(anmol_root, width=30, borderwidth=15, bd=20, justify='right', background="powder blue", foreground="black", font=font)
        e.grid(row=4, column=0, columnspan=3)

        # Create the output field
        Output = st.ScrolledText(anmol_root, height=5, width=52,bd=20, bg='#333333', fg='white', font=font)
        Output.grid(row=9, column=0, columnspan=3)

        # Function for capturing the data
        def number_input(num):
            current_value = e.get()
            e.delete(0, "end")
            e.insert(0, str(current_value) + str(num))

        # Function for clearing the entry
        def clear_values():
            e.delete(0, "end")

        # Function for enter
        def entry():
            Output.yview_moveto(1.0)
            num = e.get()
            result = game(int(num))
            e.delete(0, "end")
            Output.insert("end", result)

        # Main function
        a = random.randint(0, 100)
        def game(n):
            global count
            while count <= 9:
                if a < n:
                    count += 1
                    return f'Guess#{count-1}. You are guessing too high, guess a smaller number\n'
                elif a > n:
                    count += 1
                    return f'Guess#{count-1}.  You are guessing too low, guess a larger number\n'
                elif a == n:
                    
                    return f'You won in {count} turns.\n\n\n\nGuess another number in 10 more tries\n\n'
                    count=1
            else:
                count=1
                return f'You lose, the number was {a}\n\n\n\nYou have 10 more tries to guess another number\n\n'
        count=1    

        # Create the buttons
        Button(anmol_root, text="9", width=28, padx=20, pady=15, command=lambda: number_input(9), bg='black', fg='white', font=font).grid(row=5, column=0)
        Button(anmol_root, text="8", width=28, padx=20, pady=15, command=lambda: number_input(8), bg='black', fg='white', font=font).grid(row=5, column=1)
        Button(anmol_root, text="7", width=28, padx=20, pady=15, command=lambda: number_input(7), bg='black', fg='white', font=font).grid(row=5, column=2)
        Button(anmol_root, text="6", width=28, padx=20, pady=15, command=lambda: number_input(6), bg='black', fg='white', font=font).grid(row=6, column=0)
        Button(anmol_root, text="5", width=28, padx=20, pady=15, command=lambda: number_input(5), bg='black', fg='white', font=font).grid(row=6, column=1)
        Button(anmol_root, text="4", width=28, padx=20, pady=15, command=lambda: number_input(4), bg='black', fg='white', font=font).grid(row=6, column=2)
        Button(anmol_root, text="3", width=28, padx=20, pady=15, command=lambda: number_input(3), bg='black', fg='white', font=font).grid(row=7, column=0)
        Button(anmol_root, text="2", width=28, padx=20, pady=15, command=lambda: number_input(2), bg='black', fg='white', font=font).grid(row=7, column=1)
        Button(anmol_root, text="1", width=28, padx=20, pady=15, command=lambda: number_input(1), bg='black', fg='white', font=font).grid(row=7, column=2)
        Button(anmol_root, text="0", width=28, padx=20, pady=15, command=lambda: number_input(0), bg='black', fg='white', font=font).grid(row=8, column=1)
        Button(anmol_root, text="Clear", width=28, padx=20, pady=15, command=clear_values, bg="darkOrchid1", font=font).grid(row=8, column=0)
        Button(anmol_root, text="Enter", width=28, padx=20, pady=15, command=entry, bg='orange', fg='black', font=font).grid(row=8, column=2)

        # Run the main loop
        anmol_root.mainloop()
        pass
    elif choice == "2":
        
        # Open second program
        # Screen width and height.
        screen_width = 730
        screen_height = 355


        # Function for clearing the screen.
        def clear_screen():
            display.set("")


        # Function for displaying entered number on screen.
        def print_num(key):
            if display.get() == "ERROR":
                clear_screen()
            num = display.get() + key
            x=display.set(num)
            return(x)


        # Function for showing answer, when = is pressed.
        def get_ans():
            # Check errors in expression entered
            try:
                num = eval(display.get())
                display.set(num)
            except:
                num = "ERROR"
                display.set(num)

                
        # Creating GUI for calculator
        root = Tk()

        # Setting the geometry.
        root.geometry("{}x{}".format(screen_width, screen_height))
        root.title("Calculator")
        root.resizable(0, 0)  # Setting a fixed size for calculator.

        # Setting the font.
        text_font = tkFont.Font(family='Helvetica', size=20, weight=tkFont.BOLD)

        # Adding the display area.
        display = StringVar()
        display_area = Entry(root, width=46, bd=20, textvariable=display, state=DISABLED, justify="right", font=text_font, disabledbackground="powder blue", disabledforeground="black")                   
        display_area.grid(row=0, column=0, columnspan=4)

        # Clear button
        clr_btn = Button(root, text="CLEAR", width=43, font=text_font, bg="darkOrchid1", command=clear_screen)             
        clr_btn.grid(row=1, column=0, padx=2, sticky=W, columnspan=4)

        # Seven button
        seven_btn = Button(root, text="7", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("7"))              
        seven_btn.grid(row=2, column=0)

        # Eight button
        eight_btn = Button(root, text="8", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("8"))       
        eight_btn.grid(row=2, column=1)

        # Nine button
        nine_btn = Button(root, text="9", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("9"))            
        nine_btn.grid(row=2, column=2)

        # Divide button
        divide_btn = Button(root, text="/", width=10, font=text_font, bg="orange", fg="black", command=lambda: print_num("/"))
        divide_btn.grid(row=2, column=3)

        # Four button
        four_btn = Button(root, text="4", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("4"))
        four_btn.grid(row=3, column=0)

        # Five button
        five_btn = Button(root, text="5", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("5"))
        five_btn.grid(row=3, column=1)

        # Six button
        six_btn = Button(root, text="6", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("6"))
        six_btn.grid(row=3, column=2)

        # Mul button
        mul_btn = Button(root, text="", width=10, font=text_font, bg="orange", fg="black", command=lambda: print_num(""))
        mul_btn.grid(row=3, column=3)

        # One button
        one_btn = Button(root, text="1", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("1"))
        one_btn.grid(row=4, column=0)

        # Two button
        two_btn = Button(root, text="2", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("2"))
        two_btn.grid(row=4, column=1)

        # Three button
        three_btn = Button(root, text="3", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("3"))
        three_btn.grid(row=4, column=2)

        # Minus button
        minus_btn = Button(root, text="-", width=10, font=text_font, bg="orange", fg="black", command=lambda: print_num("-"))
        minus_btn.grid(row=4, column=3)

        # Dot button
        dot_btn = Button(root, text=".", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("."))        
        dot_btn.grid(row=5, column=0)

        # Zero button
        zero_btn = Button(root, text="0", width=10, font=text_font, bg="black", fg="white", command=lambda: print_num("0"))
        zero_btn.grid(row=5, column=1)

        # Equal button
        equal_btn = Button(root, text="=", width=10, bg="green", fg="white", font=text_font, command=get_ans)
        equal_btn.grid(row=5, column=2)

        # Plus button
        plus_btn = Button(root, text="+", width=10, font=text_font, bg="orange", fg="black", command=lambda: print_num("+"))
        plus_btn.grid(row=5, column=3)

        # Infinite loop to run the tkinter GUI
        root.mainloop() 
        pass
    
    elif choice == "3":
        # Open third program

        # Define combinations
        lowers = 'qwertyuiopasdfghjklzxcvbnm'
        uppers = 'QWERTYUIOPASDFGHJKLZXCVBNM'
        numbers = '1234567890'
        symbols = '!@#$%^&*_'
        combination = lowers + uppers + numbers + symbols

        # Create main window
        root = Tk()
        root.title("Password Generator")
        root.resizable(0, 0)

        # Set dark theme
        root.configure(bg="#2d2d2d")

        # Set font size and weight
        font = ("Helvetica", 18, "bold")

        # Create label and spinbox for password length
        length_label = Label(root,text="Enter the length of the password (between 8 and 24):", font=font, bg="#2d2d2d", fg="white")
        length_label.pack(pady=10)
        length_spinbox = Spinbox(root,from_=8, to=24, width=5, justify="center",state="readonly", font=font, disabledbackground="#2d2d2d", disabledforeground="white")
        length_spinbox.pack(pady=10)

        # Create generate button
        def generate_password():
            # Get password length from spinbox
            length = length_spinbox.get()
                # Convert length to integer and check if it is within range
            length_int = int(length)
            if 7 < length_int < 25:
                # Generate password
                password = ''.join(random.sample(combination, length_int))
                # Update password text widget with generated password
                password_text.delete("1.0", "end")
                password_text.insert("1.0", password)
                # Store generated password in global variable
                global generated_password
                generated_password = password
            else:
                password_text.delete("1.0", "end")
                password_text.insert("1.0", "The length of the password should be between 8 and 24")
        
        generate_button = Button(root,text="Generate", command=generate_password, font=font, bg="midnightblue", fg="white")
        generate_button.pack(pady=10)

        # Create copy button
        def copy_password():
            # Copy generated password to clipboard
            root.clipboard_clear()
            root.clipboard_append(generated_password)

        copy_button = Button(root,text="Copy", command=copy_password, font=font, bg="peru", fg="white")
        copy_button.pack(pady=10)

        # Create password text widget
        password_text = Text(root,width=30, height=1, font=font, bg="#2d2d2d", fg="white")
        password_text.pack(pady=10, anchor="center")

        # Run main loop
        root.mainloop()
        pass
    elif choice == "0":
        
        # Break the loop
        break
    else:
        # Print error message
        print("Enter correct input")
```
